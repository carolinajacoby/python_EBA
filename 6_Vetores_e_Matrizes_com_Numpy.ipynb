```python
np.ones(5) # Cria um array unidemensional de 1- qt é  passada no argumento
```

```python
np.ones((3,2)) # Cria um array bidimensional de zeros com x linhas e y colunas - qt é passada no argumento
```

```python
np.zeros(10)  # Cria um array unidemensional de 0- qt é  passada no argumento
```

```python
np.arange(0,123,10)  # gera um array começando em 0, indo até 122 (pois o stop = 123 não é incluído), com passo de 10

# start → valor inicial (inclusivo)

# stop → valor final (exclusivo)

# step → incremento
```

```python
np.linspace(0,123,10)

# start → valor inicial (inclusivo)

# stop → valor final (também incluso!)

# num → número total de pontos igualmente espaçados no intervalo

# Isso vai gerar 10 números igualmente espaçados entre 0 e 123 (inclusive)
```
